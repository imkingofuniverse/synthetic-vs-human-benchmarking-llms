```python
import pandas as pd

# -----------------------------
# Helper function to force 2-column reading even with extra tabs
# -----------------------------
def load_sms_file(path):
    df = pd.read_csv(
        path,
        sep="\t",
        engine="python",      # safer parser
        header=None,          # no header in original files
        names=["label", "message"],  # enforce exactly 2 columns
        usecols=[0,1],        # ignore any extra tab-split columns
        on_bad_lines="skip"   # skip corrupted rows
    )
    
    # Drop rows where label or message is missing
    df = df.dropna(subset=["label", "message"]).reset_index(drop=True)
    return df


# -----------------------------
# Load HUMAN dataset
# -----------------------------
data = load_sms_file("SMSSpamCollection")
print("Human dataset size:", len(data))


# -----------------------------
# Load SYNTHETIC datasets
# -----------------------------
df_distil = load_sms_file("Cleaned_DistillGPT2_Synthetic_SMSSpamCollection.csv")
df_gpt2   = load_sms_file("Cleaned_GPT2_Synthetic_SMSSpamCollection.csv")

print("DistilGPT-2 dataset size:", len(df_distil))
print("GPT-2 dataset size:", len(df_gpt2))


# -----------------------------
# Merge all THREE datasets
# -----------------------------
merged_df = pd.concat([data, df_distil, df_gpt2], ignore_index=True)
print(f"Total merged dataset size: {len(merged_df)}")

# -----------------------------
# Shuffle the full dataset
# -----------------------------
merged_df = merged_df.sample(frac=1, random_state=42).reset_index(drop=True)

# -----------------------------
# Save merged output
# -----------------------------
output_path = "Merged_Human_And_Synthetic_SMSSpamCollection.csv"
merged_df.to_csv(output_path, sep="\t", index=False)

print(f"Merged dataset saved to: {output_path}")
print(merged_df.head())
```
